# **Perulangan**

## **Perulangan / *Looping*** 

Python menyediakan dua statement untuk melakukan proses perulangan yaitu for dan while. Diantara kedua statement ini, secara umum for lebih banyak digunakan daripada while.

## ***For Loop***

Perulangan for digunakan untuk melakukan iterasi di dalam sequence seperti list , tuple , string , dan array atau objek yang dapat diiterasi lainnya.

**For Loop Syntax**
```
for val in sequence:
    loop body
```
Pada proses iterasi ini, val adalah variabel yang diambil dari elemen-elemen di dalam sequence untuk setiap iterasi.

Perulangan akan berlangsung hingga elemen terakhir dari sequence. Pada bagian body dari for loop
diletakkan dalam indentasi.

**For Loop Flowchart**

Screenshot (47).png

**Example: Python for Loop**

````python
# Program untuk menghitung jumlah semua angka dalam sebuah list

# List angka
list_angka = [6, 5, 3, 8, 4, 2, 5, 4, 11]

# Variabel untuk menyimpan hasil penjumlahan
jumlahan = 0

# Iterasi di dalam list
for angka in list_angka:
  jumlahan = jumlahan+angka
  
print("Jumlah :", jumlahan)
````

**The range() function**

Kita dapat membuat sebuah sequence angka menggunakan fungsi range(). range(10) akan menghasilkan sequence angka dari 0 sampai 9 (10 angka).

Kita juga dapat mendefinisikan nilai awal, nilai akhir dan step sizenya yaitu range(start, stp, step_size). Secara default step_size bernilai 1 jika tidak ditentukan secara khusus.

Kita dapat menggunakan fungsi range() di dalam perulangan for untuk melakukan iterasi di dalam sequence numbers . Ini dapat kita kombinasikan dengan fungsi len() untuk melakukan iterasi melalui index dari sebuah sequence. 

Berikut contohnya:

````python
# Program untuk melakukan iterasi di dalam list melalui indeks
musik = ['dangdut', 'campursari', 'keroncong']

# Iterasi di dalam list melalui indeks
for i in range(len(musik)):
  print("Aku suka", musik[i])
````

**for loop with else**

Sebuah for loop dapat memiliki opsi blok else. Blok else akan dijalankan jika iterasi telah selesai dilakukan dalam sebuah sequence.

Keyword break digunakan untuk menghentikan iterasi. Dalam kasus itu maka bagian else diabaikan.

Sehingga, bagian iterasi else hanya dijalankan jika tidak terjadi break.

Berikut adalah beberapa contoh sebagai ilustrasi:

````python
digits = [0, 1, 5]

for i in digits: 
  print(i)
else:
  print("Tidak ada elemen yang tersisa.")
````

Ketika iterasi berjalan maka akan dilakukan print item hingga iterasi selesai. Ketika iterasi pada sequence selesai baru perintah pada blok else dijalankan yaitu print Tidak ada elemen yang tersisa.

Statement for ... else dapat digunakan bersama dengan keyword break untuk menjalankan blok else hanya ketika break tidak dieksekusi. Lihat contoh berikut:

````python
# Program untuk menampilkan nilai tugas siswa.
nama_siswa = 'Butet'
daftar_nilai = {'Bambang': 90, 'Andri': 55, 'Candra': 77}

for siswa in daftar_nilai:
  if siswa == nama_siswa:
    print(daftar_nilai[siswa])
    break
else:
  print('Tidak ada data nilai atas nama', nama_siswa, '!')
````

## ***While Loop***

While loop pada Python digunakan untuk melakukan iterasi pada blok kode selama nilai dari test expression adalah true.

Looping ini biasanya digunakan ketika kita tidak tau berapa kali iterasi akan dilakukan hingga sebuah kondisi yang diinginkan dapat dicapai.

**Syntax of while Loop in Python**
```
while test_expression:
    Body of while
```

Pada loop while, test expression justru dicheck pertama kali. Kemudian statement - statement di dalam body of while akan dieksekusi jika nilai dari test expression adalah true. Setelah satu iterasi dijalankan, test expression dicek kembali. Proses ini terus berlanjut hingga nilai evaluasi dari test expression adalah False.

Pada Python, bagian body pada loop while dituliskan menggunakan indentasi.

Sebagai catatan tambahan bahwa Python menginterpretasikan non-zero values sebagai True. None dan 0 diinterpretasikan sebagai False.

**While Loop Flowchart**

Screenshot (49).png

**Example: Python while Loop**

````python
# Program untuk menjumlahkan semua
# Bilangan asli hingga suku ke-n
# Jumlahan = 1+2+3+...+n

n = 10
# Nilai awal penjumlahan dan iterasi
jumlahan = 0
i = 1

while i <= n:
  jumlahan = jumlahan + i
  i = i+1 # Update iterasi

# Cetak hasil penjumlahan
print("jumlahan =", jumlahan)
````

Pada program di atas, test expression akan bernilai True selama nilai iterasi i kurang dari atau sama dengan n.

Kita perlu mengupdate nilai dari variabel pencatat iterasi i agar iterasi tidak berlangsung secara terus menerus(infinity loop).


**While loop with else**

Sama dengan for loop kita juga dapat mengkombinasikan while loop dengan blok else.

Bagian else akan dieksekusi jika nilai pada test expression pada pada while loop bernilai False.

Perhatikan contoh berikut:

````python
counter = 0
while counter < 3:
  print("Di dalam Loop")
  counter = counter + 1
else:
  print("Di dalam else")
````

1. Buatlah fungsi untuk menghitung diskon suatu pembayaran belanjaan dengan input berupa dictionary untuk daftar harga dan dictionary untuk daftar belanjaan!

Contoh:

```
daftar_harga = {'telur/kg':20000, 'gula_pasir/kg':16000, 'teh/bks':4000}
daftar_belanjaan = {'telur':2, 'gula_pasir':5}

jumlah_bayar = pembayaran(daftar_harga, daftar_belanjaan)
```
Output

    120000

````python
daftar_harga = {'telur' : 20000, 'gula_pasir' : 16000, 'teh' : 4000}
daftar_belanjaan = {'telur' : 2, 'gula_pasir' : 5, 'teh' : 10}
def pembayaran(daftar_harga, daftar_belanjaan):
  jumlah = 0
  for barang in daftar_harga:
    for beli in daftar_belanjaan:
      if barang == beli:
        jumlah += daftar_harga[beli]*daftar_belanjaan[beli]
  disc = jumlah-jumlah*0.2
  return disc
jumlah_bayar = pembayaran(daftar_harga, daftar_belanjaan)
print(jumlah_bayar)
````

## **IF Statements**

**Python IF Statement Syntax**
```
if test expression:
   statement(s)
```
Program akan mengevaluasi nilai dari test expression yang diberikan kemudian akan mengeksekusi statement(s) hanya jika test expression bernilai True.

Jika test expression bernilai False , maka statement(s) tidak dieksekusi.

Seperti telah disampaikan pada pertemuan sebelumnya bahwa bagian body dari statement if diindikasikan dengan indentasi. Statement-statement yang akan dijalankan jika test expression bernilai True ditulis pada baris selanjutnya dan menjorok ke dalam.

Sebagai catatan tambahan bahwa Python menginterpretasikan non-zero values sebagai True. None dan 0 diinterpretasikan sebagai False.

**Python IF Statement Flowchart**

Screenshot (51).png


**Example: Python IF Statement**

````python
# Jika angka adalah bilangan positif maka print "adalah bilangan positif"
angka = 3
if angka > 0:
  print(angka, "adalah bilangan positif.")
print("Pernyataan ini selalu ditampilkan.")

angka = -1
if angka > 0:
  print(angka, "adalah bilangan positif.")
print("Pernyataan ini juga selalu ditampilkan.")
````

Pada contoh di atas, angka > 0 adalah test expression. Bagian body dari if dieksekusi hanya jika hasil evaluasi dari test expression adalah bernilai True.

Ketika variabel angka bernilai 3, test expression bernilai True sehingga statement di dalam bagian body dari if akan dieksekusi. Sedangkan ketika variabel angka bernilai -1, test expression bernilai False sehingga statement di dalam bagian body dari if tidak dieksekusi(di skip).

Statement print() di luar blok if (yang tidak terindentasi) akan selalu dieksekusi.

## **IF ELSE Statements**

**Python IF ELSE Statement Syntax**

```
if test expression:
   Body of if
else :
   Body of else
```

Statement if ... else mengevaluasi test expression dan akan mengeksekusi bagian body of if hanya jika test expression bernilai True.

Jika test expression bernilai False , maka bagian body of else akan dieksekusi. Indentasi juga digunakan pada bagian ini.

**Python IF ELSE Statement Flowchart**

Screenshot (53).png

**Example: Python IF ELSE Statement**

````python
# Program disusun untuk mengecek apakah angka merupakan bilangan positif atau negatif
# dan menampilkan tulisan yang sesuai

angka = -5

# Kamu bisa mencoba angka-angka lainnya berikut:
# angka = -5
# angka = 0

if angka >= 0:
  print("Bilangan Positif atau Nol")
else:
  print("Bilangan Negatif")
````

Pada contoh di atas, ketika angka sama dengan 3, test expression bernilai True dan bagian body dari if dieksekusi sedangkan bagian body dari else diabaikan.

Ketika angka sama dengan -5, test expression bernilai False dan bagian body dari if diabaikan sedangkan bagian body dari else dieksekusi.

Jika angka sama dengan 0, test expression bernilai True dan bagian body dari if dieksekusi
sedangkan bagian body dari else diabaikan.

## **Python Nested IF Statements**

Kita dapat membuat statement if...elif...else di dalam statement if...elif...else lainnya. Ini disebut sebagai nesting dalam computer programming.

Sejumlah pernyataan ini dapat bersarang (nesting) di dalam satu sama lain. Indentasi adalah satu-satunyacara untuk mengetahui tingkat bersarang. Mereka bisa membingungkan, jadi mereka harus dihindari kecuali jika perlu.

**Example: Python NESTED IF Statement**

````python
# Program disusun untuk mengecek apakah angka merupakan bilangan positif, negatif atau nol
# dan menampilkan tulisan yang sesuai

angka = float(input("Masukkan angka: "))
if angka >= 0:
  if angka == 0:
    print("Nol")
  else:
    print("Bilangan Positif")
else:
  print("Bilangan Negatif")
````

2. Diberi sebuah string, tulislah sebuah fungsi python untuk memeriksa apakah string tersebut merupakan
palindrom atau bukan. Sebuah string dikatakan palindrom jika susunan terbalik dari string tersebut sama
dengan string pada susunan awalnya. Misalnya, "radar" adalah palindrom, tetapi "radix" bukanlah
palindrom.

````python
a = input("Kata: ")
z = len(a)
kondisi = True
for i in range (0, z):
  if (kondisi):
    b = z-(i+1)
    if a[i] is a[b]:
      kondisi = True
    else:
      kondisi = False
if (kondisi):
  print('Palindrom')
else:
  print('Bukan Palindrom')
````
